```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

# Model evaluation data for BBC News (Small and medium document size)
data = {
    'Model': [
        'TF-IDF', 'TextRank', 'LexRank', 
        'SBERT + KMeans (3)', 'SBERT + KMeans (5)', 
        'SBERT + LDA', 'SBERT + TextRank'
    ],
    'ROUGE-1': [0.4520, 0.6389, 0.5843, 0.3808, 0.3591, 0.4455, 0.4532],
    'ROUGE-2': [0.3468, 0.5357, 0.4615, 0.2519, 0.2325, 0.3513, 0.3538],
    'ROUGE-L': [0.4418, 0.6285, 0.5719, 0.3594, 0.3415, 0.4380, 0.4449],
    'Cosine Similarity': [0.7598, 0.8546, 0.8315, 0.6935, 0.7038, 0.8475, 0.8509],
    'Precision': [0.3019, 0.2807, 0.2852, 0.3319, 0.2620, 0.2904, 0.3064],
    'Recall': [0.7769, 0.7359, 0.6882, 0.5007, 0.6075, 0.7984, 0.7730],
    'F1 Score': [0.4018, 0.3772, 0.3723, 0.3627, 0.3349, 0.3977, 0.4049]
}

df = pd.DataFrame(data)

# Plot settings
metrics = ['ROUGE-1', 'ROUGE-2', 'ROUGE-L', 'Cosine Similarity', 'Precision', 'Recall', 'F1 Score']

colors = ['blue', 'green', 'red']
# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(14, 18))
metric_sets = [metrics[:3], metrics[3:4], metrics[4:]]

for i, metric_group in enumerate(metric_sets):
    df_plot = df.set_index('Model')[metric_group]
    df_plot.plot(kind='bar', ax=axes[i], color=colors)
    axes[i].set_title(f'Summarization Evaluation Metrics: {" & ".join(metric_group)}')
    axes[i].set_ylabel('Score')
    axes[i].set_xlabel('Model')
    axes[i].legend(title='Metric')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
